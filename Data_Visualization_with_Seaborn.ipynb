```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import numpy as np

df = pd.read_excel("/content/MoonMindAI_Matplotlib_practise.xlsx")

sns.set_theme(style="whitegrid", context="notebook")

plt.figure(figsize=(13, 6))

sns.pointplot(
    data=df,
    x="Risk_Level",
    y="Reward_Score",
    hue="Mission_State",
    order=["Low", "Medium", "High"],
    hue_order=["Safe", "Unsafe"],
    estimator=np.mean,
    errorbar=None,
    palette="Set2",
    markers=["o", "s"],
    linestyles=["-", "--"],
    dodge=0.25,
    capsize=0.1,
    legend=True
)

plt.title("MoonMind AI: Average Reward Score by Risk Level", fontsize=16, fontweight="bold")
plt.xlabel("Risk Level", fontsize=12)
plt.ylabel("Average Reward Score", fontsize=12)
plt.grid(True, linestyle="--", alpha=0.4)
plt.legend(title="Mission State")

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

df = pd.read_excel("/content/MoonMindAI_Matplotlib_practise.xlsx")

numeric_df = df[
    [
        "Altitude",
        "Fuel",
        "Fuel_Used",
        "Landing_Error",
        "Distance",
        "Slope_Degree",
        "Terrain_Roughness",
        "Reward_Score"
    ]
]

corr_data = numeric_df.corr()

sns.set_theme(style="white", context="notebook")

plt.figure(figsize=(11, 7))

sns.heatmap(
    corr_data,
    annot=True,
    fmt=".2f",
    cmap="coolwarm",
    center=0,
    linewidths=0.7,
    linecolor="white",
    cbar=True,
    square=True
)

plt.title("MoonMind AI: Correlation Heatmap", fontsize=16, fontweight="bold")

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

df = pd.read_excel("/content/MoonMindAI_Matplotlib_practise.xlsx")

sns.set_theme(style="whitegrid", context="notebook")

sns.pairplot(
    data=df,
    vars=[
        "Fuel",
        "Fuel_Used",
        "Landing_Error",
        "Slope_Degree",
        "Terrain_Roughness",
        "Reward_Score"
    ],
    hue="Mission_State",
    hue_order=["Safe", "Unsafe"],
    palette="Set2",
    kind="scatter",
    diag_kind="hist",
    markers=["o", "s"],
    height=2.2,
    aspect=1,
    corner=True,
    plot_kws={
        "alpha": 0.8,
        "s": 55,
        "edgecolor": "black",
        "linewidth": 0.4
    },
    diag_kws={
        "alpha": 0.7,
        "bins": 8
    }
)

plt.suptitle("MoonMind AI: Pair Plot of Mission Metrics", y=1.02, fontsize=16, fontweight="bold")
plt.show()
```
